1. Print numbers 1 to 5

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

2. Sum of numbers 1 to 10

```python
total = 0
i=1
while i <= 10:
    total += i
    i += 1
print(total)
```

3. Multiplication table of 3

```python
i = 1
while i <= 10:
    print("",5*i)
    i += 1
```

4. Print even numbers from 1 to 20

```python
i = 1
while i <= 20:
    if i % 2 == 0:
        print(i)
    i += 1
```

5. . Factorial of a number

```python
num = 5
fact = 1
while num > 0:
    fact *= num
    num -= 1
print(fact)
```
